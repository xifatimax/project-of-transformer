###  $\color{#FF1493}{\text{Table1:}}$ $\color{#1F1F1F}{\text{Vp>Vs, ( step down)}}$

```python
import pandas as pd
import numpy as np
data={'Vp_in_V':[1,2,3,4,5,6],'Vs_in_V':[0.40,0.9,1.4,1.9,2.4,2.9]}
df=pd.DataFrame(data)
df.style.set_properties(**{'background-color': '#36648B', 'color': 'white'})
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set()
x=df.Vs_in_V
y=df.Vp_in_V
m,b=np.polyfit(x,y,1)
plt.plot(x,y,'o',color='black')
plt.plot(x,m*x+b,color='#36648B')
plt.xlabel('Vs(V)',fontsize=15)
plt.ylabel('Vp(V)',fontsize=15)
plt.title('Primery Voltage & Secondary Voltage',fontsize=22,color='#36648B')
plt.show()
```

###  $\color{#36648B}{\text{Find slope:}}$ 


```python
import numpy as np
x=[1.5,1]
y=[3.1,2.1]
slope,intercept = np.polyfit(x,y,1)
print('The slope is=',round(slope,1))
```

# A particle moves along the x axis according to 
# the equation x = 2.00 + 3.00t - 1.00t2 , where x is in meters and t is in seconds . At t = 3.00 s , find: 
### ( a ) the position of the particle.
### ( b ) its velocity.
### ( c ) its acceleration.

## $\color{#8C7698}{\text{( a ) the position of the particle.}}$

```python
x=lambda t : 2+3*t-1*t**2
position=x(3)
print( 'Position at t=3:',position,'m')
```

## $\color{#8C7698}{\text{( b ) its velocity.}}$

```python
import sympy as sym
from sympy import symbols
t=symbols('t')
x=2+3*t-1*t**2
v=x.diff(t)
print('dx/dt:',v)
```

```python
t=3
v= 3 - 2*t
print('Velocity at t=3:',v,'m/s')
```

## $\color{#8C7698}{\text{( c ) its acceleration.}}$

```python
import sympy as sym
from sympy import symbols
t=symbols('t')
x=2+3*t-1*t**2
a=x.diff(t,2)
print('d^2x/dt^2:',a)
```

```python
t=3
a=-2
print('Acceleration at t=3:',a,'m/s^2')
```
